```python
from datetime import date

def calculate_domain_age(creation_date_str, expiry_date_str):

  try:
    creation_date = date.fromisoformat(creation_date_str)
    today = date.today()
    if expiry_date_str:
      expiry_date = date.fromisoformat(expiry_date_str)
      
      if expiry_date < today:
        return "Domain Expired"
    
    age_in_days = (today - creation_date).days
    return age_in_days / 365
  except ValueError:
    return "Invalid date format. Please use YYYY-MM-DD."


creation_date = "1935-01-01"
expiry_date = "2025-12-31" 

domain_age = calculate_domain_age(creation_date, expiry_date)
print(f"The domain is approximately {domain_age:.2f} years old.")
```
